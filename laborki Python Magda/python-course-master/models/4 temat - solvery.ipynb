```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def F(x, t):
    dx = [0, 0]
    dx[0] = x[1]
    dx[1] = -x[0]-0.5*x[1]
    return dx

t_min = 0
t_max = 20 
h = 0.01
t = np.arange(t_min, t_max+h, h)

initial_x = ((1,0))

X = odeint(F, initial_x, t)
plt.figure(1)
plt.plot(t,X)
plt.figure(2)
plt.plot(X[:,0],X[:,1])
plt.axis('equal')
plt.show()
```

```python
knots = np.linspace(-2,2,16)
X, Y = np.meshgrid(knots, knots)
dX, dY = (1/2)*X-Y-(1/2)*(X**3+X*Y**2), X+(1/2)*Y-(1/2)*(Y**3+Y*X**2)
plt.figure(1)
plt.quiver(X, Y, dX, dY)
plt.figure(2)
plt.streamplot(X, Y, dX, dY, density=1)
plt.show()
```

```python
knotsX = np.linspace(-8,8,16)
knotsY = np.linspace(-3,3,16)
X, Y = np.meshgrid(knotsX, knotsY)
#dX, dY = (1/2)*X-Y-(1/2)*(X**3+X*Y**2), X+(1/2)*Y-(1/2)*(Y**3+Y*X**2)
dX, dY = Y,-np.sin(X) -0.5*Y
plt.figure(1)
plt.quiver(X, Y, dX, dY)
plt.figure(2)
plt.streamplot(X, Y, dX, dY, density=1.5)
plt.show()
```

```python
# Fitzhugh-Nagumo
b = 0.9
a = (1/2) + (1/6)*b
I = 0.4

knots = np.linspace(-4,4,16)
X, Y = np.meshgrid(knots, knots)
dX, dY = Y + X - (1/3)*X**3 + I, -X + a - b*Y
plt.figure(1)
plt.quiver(X, Y, dX, dY)
plt.figure(2)
plt.streamplot(X, Y, dX, dY, density=1)
plt.show()
```

```python
from numpy import array, arange, linspace, meshgrid, zeros
from matplotlib.pyplot import plot, show, streamplot, figure
from scipy.integrate import odeint

def F(x, t):
    dx = zeros(2)
    dx[0] = x[1]
    dx[1] = -x[0] - 0.8*x[1]
    return dx

T = 30
h = 0.01

t = arange(0,T,h)
x0 = array([1,0])
x = odeint(F,x0,t)

figure(1)
plot(t,x[:,0],'k')
plot(t,x[:,1],'g')

figure(2)
knotsX1 = linspace(-1,1,16)
knotsX2 = linspace(-1,1,16)
X1, X2 = meshgrid(knotsX1, knotsX2)
dX1 = zeros(X1.shape)
dX2 = zeros(X2.shape)
for i in range(X1.shape[0]):
    for j in range(X1.shape[1]):
        dX = F([X1[i,j],X2[i,j]],-1)
        dX1[i,j] = dX[0]
        dX2[i,j] = dX[1]

streamplot(X1, X2, dX1, dX2)
show()
```
