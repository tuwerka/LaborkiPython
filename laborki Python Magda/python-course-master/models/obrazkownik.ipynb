```python
from numpy import zeros, array
from matplotlib.pyplot import show, imshow, axis, figure
from matplotlib.image import imread

from numpy.random import randint

image = imread('cat.jpg')

r = array(image[:,:,0])
g = array(image[:,:,1])
b = array(image[:,:,2])


for i in range(100000):
    ma = [randint(0,r.shape[0]), randint(0,r.shape[1])]
    mb = [randint(0,r.shape[0]), randint(0,r.shape[1])]
    pom = r[ma[0],ma[1]]
    r[ma[0],ma[1]] = r[mb[0],mb[1]]
    r[mb[0],mb[1]] = pom
    pom = g[ma[0],ma[1]]
    g[ma[0],ma[1]] = g[mb[0],mb[1]]
    g[mb[0],mb[1]] = pom
    pom = b[ma[0],ma[1]]
    b[ma[0],ma[1]] = b[mb[0],mb[1]]
    b[mb[0],mb[1]] = pom


            
new_image = zeros(image.shape, dtype='uint8')
new_image[:,:,0] = r
new_image[:,:,1] = g
new_image[:,:,2] = b

figure(figsize=(10,10))
imshow(new_image)
axis('off')
show()
```

```python
from numpy import zeros, array, mean, uint8, median
from matplotlib.pyplot import show, imshow, axis, figure
from matplotlib.image import imread

from numpy.random import randint

image = imread('cat.jpg')

r = array(image[:,:,0])
g = array(image[:,:,1])
b = array(image[:,:,2])


r = r.astype(float)
g = g.astype(float)
b = b.astype(float)

nr = r
ng = g
nb = b

s = 8

for i in range(s,r.shape[0]-s):
    for j in range(s,r.shape[1]-s):
        nr[i,j] = median(r[i-s:i+s+1,j-s:j+s+1])
        ng[i,j] = median(g[i-s:i+s+1,j-s:j+s+1])
        nb[i,j] = median(b[i-s:i+s+1,j-s:j+s+1])

r = nr.astype(uint8)
g = ng.astype(uint8)
b = nb.astype(uint8)


            
new_image = zeros(image.shape, dtype='uint8')
new_image[:,:,0] = r
new_image[:,:,1] = g
new_image[:,:,2] = b

figure(1, figsize=(10,10))
imshow(image)
axis('off')
figure(2, figsize=(10,10))
imshow(new_image)
axis('off')
show()
```

```python
from numpy import zeros, array
from matplotlib.pyplot import show, imshow, axis, figure
from matplotlib.image import imread

from numpy.random import randint

image = imread('cat.jpg')

r = array(image[:,:,0])
g = array(image[:,:,1])
b = array(image[:,:,2])


for i in range(int(r.shape[0]/2)):
    for j in range(r.shape[1]):
        ma = [i, j]
        mb = [r.shape[0]-i-1, j]
        pom = r[ma[0],ma[1]]
        r[ma[0],ma[1]] = r[mb[0],mb[1]]
        r[mb[0],mb[1]] = pom
        pom = g[ma[0],ma[1]]
        g[ma[0],ma[1]] = g[mb[0],mb[1]]
        g[mb[0],mb[1]] = pom
        pom = b[ma[0],ma[1]]
        b[ma[0],ma[1]] = b[mb[0],mb[1]]
        b[mb[0],mb[1]] = pom

for i in range(r.shape[0]):
    for j in range(int(r.shape[1]/2)):
        ma = [i, j]
        mb = [i, r.shape[1]-j-1]
        pom = r[ma[0],ma[1]]
        r[ma[0],ma[1]] = r[mb[0],mb[1]]
        r[mb[0],mb[1]] = pom
        pom = g[ma[0],ma[1]]
        g[ma[0],ma[1]] = g[mb[0],mb[1]]
        g[mb[0],mb[1]] = pom
        pom = b[ma[0],ma[1]]
        b[ma[0],ma[1]] = b[mb[0],mb[1]]
        b[mb[0],mb[1]] = pom

            
new_image = zeros(image.shape, dtype='uint8')
new_image[:,:,0] = r
new_image[:,:,1] = g
new_image[:,:,2] = b

figure(figsize=(10,10))
imshow(new_image)
axis('off')
show()
```
