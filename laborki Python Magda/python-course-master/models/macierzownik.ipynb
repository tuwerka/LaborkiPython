```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[0, 0, 1], [1, 1, 0], [0, 1, 0]])
plt.imshow(x, interpolation='nearest', cmap=plt.cm.binary)   #bilinear, bicubic
plt.show()
```

```python
delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

im = plt.imshow(Z, interpolation='bilinear', cmap=plt.cm.RdYlGn,
                origin='lower', extent=[-3, 3, -3, 3],
                vmax=abs(Z).max(), vmin=-abs(Z).max())

plt.show()
```

```python
A = np.arange(0,25).reshape(5,5)
B = np.copy(A)

B[4,:] = A[1,:]

plt.figure()
plt.subplot(1,2,1)
plt.imshow(A)
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(B)
plt.colorbar(extend='both')
plt.clim(0, 24)

plt.show()
```
