```python
import numpy as np
import matplotlib.pyplot as plt

# parametry symulacji
T = 5
h = 0.1

# parametry modelu: x' = ax
a = -1
initial_x = 2

t = np.arange(0,T,h)
x = np.zeros(t.shape)
x[0] = initial_x

for i in range(t.size-1):
    x[i+1] = x[i] + h*(-x[i])

plt.plot(t,x,'o')
plt.xlabel('t', fontsize=14)
plt.ylabel('x', fontsize=14)
plt.show()
```

```python
# parametry symulacji
T = 20
h = 0.1

# parametry modelu x'' + ax' + bx = 0
a, b = 0.5, 1
initial_x, initial_y = 1.0, 0.0

t = np.arange(0,T,h)
x = np.zeros(t.shape)
y = np.zeros(t.shape)
x[0], y[0] = initial_x, initial_y

for i in range(t.size-1):
    x[i+1] = x[i] + h*(y[i])
    y[i+1] = y[i] + h*(-b*x[i] - a*y[i])

plt.plot(t,x,'k', label='x')
plt.plot(t,y,'g', label='y')
plt.xlabel('t', fontsize=14)
plt.ylabel('stan', fontsize=14)
plt.legend(loc='upper right', fontsize=14)
plt.show()

# pokazać, że dla a = 0 (wyłączone tarcie) ujawnia się słabość schematu Eulera i z ratunkiem przychodzi skorygowany schemat Eulera
```

```python
# skorygowany schemat Eulera
T = 20
h = 0.1

# parametry modelu x'' + ax' + bx = 0
a, b = 0, 1
initial_x, initial_y = 1.0, 0.0

t = np.arange(0,T,h)
x = np.zeros(t.shape)
y = np.zeros(t.shape)
x[0], y[0] = initial_x, initial_y

for i in range(t.size-1):
    x[i+1] = x[i] + h*(y[i])
    y[i+1] = y[i] + h*(-b*x[i+1] - a*y[i])

plt.plot(t,x,'k', label='x')
plt.plot(t,y,'g', label='y')
plt.xlabel('t', fontsize=14)
plt.ylabel('stan', fontsize=14)
plt.legend(loc='upper right', fontsize=14)
plt.show()

# pokazać, że dla a = 0 ujawnia się słabość schematu Eulera i z ratunkiem przychodzi skorygowany schemat Eulera
```

```python
T = 100
h = 0.01

# parametry modelu x'' + ax' + bx = 0
a, b, tau = 0.7, 0.8, 10
initial_x, initial_y = 0.5, 0.3

t = np.arange(0,T,h)
x = np.zeros(t.shape)
y = np.zeros(t.shape)
x[0], y[0] = initial_x, initial_y

for i in range(t.size-1):
    x[i+1] = x[i] + h*(-y[i] + x[i] - x[i]**3/3 + 0.5)
    y[i+1] = y[i] + h*(x[i+1] + a - b*y[i])/tau

plt.figure(1)
plt.plot(t,x,'k', label='x')
plt.plot(t,y,'g', label='y')
plt.xlabel('t', fontsize=14)
plt.ylabel('stan', fontsize=14)
plt.legend(loc='upper right', fontsize=14)
plt.figure(2)
plt.plot(x,y,color='orange')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 80
h = 0.01

# parametry modelu x'' + ax' + bx = 0
a, b, c = 0.2, 0.2, 5.7
initial_x, initial_y, initial_z = 0.5, 0.3, 0.8

t = np.arange(0,T,h)
x = np.zeros(t.shape)
y = np.zeros(t.shape)
z = np.zeros(t.shape)
x[0], y[0], z[0] = initial_x, initial_y, initial_z

for i in range(t.size-1):
    x[i+1] = x[i] + h*(-y[i] - z[i])
    y[i+1] = y[i] + h*(x[i+1] + a*y[i])
    z[i+1] = z[i] + h*(b + z[i] * (x[i+1] - c))

plt.figure(1)
plt.plot(t,x,'k', label='x')
plt.plot(t,y,'g', label='y')
plt.plot(t,z,'b', label='z')
plt.xlabel('t', fontsize=14)
plt.ylabel('stan', fontsize=14)
plt.legend(loc='upper right', fontsize=14)
plt.show()
```

```python
# parametry symulacji
T = 40
h = 0.01

mi, om = 2, 0.7
initial_x, initial_y = 1, -1

t = np.arange(0,T,h)
x = np.zeros(t.shape)
y = np.zeros(t.shape)
x[0], y[0] = initial_x, initial_y

for i in range(t.size-1):
    x[i+1] = x[i] + h*(y[i])
    y[i+1] = y[i] + h*(mi*(1-x[i+1]**2)*y[i] - om**2*x[i+1])

plt.plot(t,x,'k', label='x')
plt.plot(t,y,'g', label='y')
plt.xlabel('t', fontsize=14)
plt.ylabel('stan', fontsize=14)
plt.legend(loc='upper right', fontsize=14)
plt.show()
```
