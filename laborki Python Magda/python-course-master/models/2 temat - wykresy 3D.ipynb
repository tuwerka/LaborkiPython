```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_knots = np.linspace(-3,3,41)
y_knots = np.linspace(-3,3,41)
X, Y = np.meshgrid(x_knots, y_knots)
Z = X**2 + X*Y**2 + Y**3
ax = Axes3D(plt.figure())
ax.plot_surface(X, Y, Z, cmap=plt.cm.rainbow)
plt.show()
```

```python
x_knots = np.linspace(-3*np.pi,3*np.pi,201)
y_knots = np.linspace(-3*np.pi,3*np.pi,201)
X, Y = np.meshgrid(x_knots, y_knots)
R = np.sqrt(X**2+Y**2)
Z = np.cos(R)**2*np.exp(-0.3*R)
ax = Axes3D(plt.figure(figsize=(8,5)))
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0.4)
plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
t = np.linspace(0,5*np.pi, 501)
ax.plot(np.cos(t), np.sin(t), t)
plt.show()
```

```python
plt.figure()
x_values = np.linspace(-3, 3, 41)
y_values = np.linspace(0, 10, 21)
X, Y = np.meshgrid(x_values, y_values)
Z = np.cos(X) + np.sin(Y)
cs = plt.contour(X, Y, Z, linewidths=3, colors='k')
plt.clabel(cs,fontsize=10)
plt.show()
```

```python
import numpy.random as rnd

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 400

x1 = 6*rnd.rand(n)-3
y1 = 6*rnd.rand(n)-3
z1 = 6*rnd.rand(n)-3

x2 = rnd.randn(n)+10
y2 = rnd.randn(n)
z2 = rnd.randn(n)

ax.scatter(x1, y1, z1, c='b', marker='^')
ax.scatter(x2, y2, z2, c='y', marker='+')

ax.set_xlim3d(-3,13)
ax.set_ylim3d(-4,4)
ax.set_zlim3d(-4,4)

# ax.axis('equal')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
